```python
from dash import Dash, html, dcc
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# تحميل البيانات
try:
    day = pd.read_csv('day_wise.csv')  # تأكد من مسار الملف
    day['Date'] = pd.to_datetime(day['Date'])  # تحويل العمود Date إلى صيغة تاريخ
    day.rename(columns={'type': 'case_type'}, inplace=True)  # تجنب التعارض مع الكلمة المحجوزة
except Exception as e:
    print(f"Error loading data: {e}")
    day = pd.DataFrame(columns=['Deaths', 'Date', 'Active', 'case_type'])

# إنشاء التطبيق
app = Dash()

# تصميم الواجهة
app.layout = html.Div(children=[
    html.H1(children='Dashboard for COVID-19 Historical Data'),

    # القائمة المنسدلة لاختيار الوفيات
    dcc.Dropdown(
        id='deaths-dropdown',
        options=[{'label': i, 'value': i} for i in day['Deaths'].unique()],
        value=day['Deaths'].iloc[0] if not day.empty else None,
        placeholder="Select Death Count"
    ),
    
    # الرسومات البيانية
    dcc.Graph(id='line-graph'),
    dcc.Graph(id='bar-chart'),
    dcc.Graph(id='pie-chart')
])

# الوظيفة التفاعلية لتحديث الرسم البياني الأول (خطّي)
@app.callback(
    Output('line-graph', 'figure'),
    [Input('deaths-dropdown', 'value')]
)
def update_line_graph(selected_deaths):
    if day.empty or selected_deaths is None:
        return px.line(title="No data available")
    
    filtered_day = day[day['Deaths'] == selected_deaths]
    line_fig = px.line(filtered_day,
                       x='Date', y='Active',
                       color='case_type',
                       title=f'Active Cases for Death Count: {selected_deaths}')
    return line_fig

# الوظيفة التفاعلية لتحديث الرسم البياني الثاني (عمودي)
@app.callback(
    Output('bar-chart', 'figure'),
    [Input('deaths-dropdown', 'value')]
)
def update_bar_chart(selected_deaths):
    if day.empty or selected_deaths is None:
        return px.bar(title="No data available")
    
    filtered_day = day[day['Deaths'] == selected_deaths]
    bar_fig = px.bar(filtered_day,
                     x='Date', y='Active',
                     color='case_type',
                     title=f'Active Cases Bar Chart for Death Count: {selected_deaths}')
    return bar_fig

# الوظيفة التفاعلية لتحديث الرسم البياني الثالث (دائري)
@app.callback(
    Output('pie-chart', 'figure'),
    [Input('deaths-dropdown', 'value')]
)
def update_pie_chart(selected_deaths):
    if day.empty or selected_deaths is None:
        return px.pie(title="No data available")
    
    filtered_day = day[day['Deaths'] == selected_deaths]
    pie_fig = px.pie(filtered_day,
                     names='case_type', values='Active',
                     title=f'Case Distribution by Type for Death Count: {selected_deaths}')
    return pie_fig

# تشغيل الخادم
if __name__ == '__main__':
    app.run_server(debug=True, host='127.0.0.1', port=8050)
    host = '127.0.0.1'  # عنوان IP المحلي
    port = 8050         # المنفذ الافتراضي
    print(f"App is running at http://{host}:{port}/")
    app.run_server(debug=True, host=host, port=port)
```
